```python
import random
def unico(numero,lista):#En esta primera parte del programa, he incluido una mejora que hace la comprobación para que cuando se generen los números en la lista, no aparezcan repetidos
  num_unico=True
  for i in range(len(lista)):
    if numero==lista[i]:
      num_unico=False
      break
  return num_unico
lista=[]
contador=0
while contador<100:
  numero=random.randint(0,100)
  if unico(numero,lista):
    lista.append(numero)
    contador = contador +1 
print(lista)
lista.sort()
print(lista)
```

```python
# Escogemos la lista desordenada generada en el ejercicio 1.
numeros=[28, 61, 53, 5, 60, 23, 47, 39, 18, 59, 64, 88, 52, 34, 57, 8, 30, 40, 44, 94, 20, 65, 54, 27, 79, 9, 97, 24, 45, 19, 0, 13, 51, 56, 77, 17, 50, 84, 29, 58, 75, 1, 87, 16, 72, 55, 62, 96, 68, 66, 85, 78, 6, 10, 67, 38, 89, 43, 92, 4, 31, 12, 80, 86, 11, 2, 71, 15, 90, 98, 33, 48, 91, 25, 99, 46, 21, 93, 83, 100, 63, 74, 3, 41, 49, 32, 26, 95, 76, 7, 70, 36, 14, 35, 42, 81, 37, 69, 22, 82]
# 1. Función para añadir un número a la lista
def añadir_numero():
    numero = int(input("Introduce un número: "))
    numeros.append(numero)
    print("Se ha introducido el número correctamente:")
    print(numeros)

# 2. Función para eliminar un número según su posición
def eliminar_numero():
    posicion = int(input("Introduce la posición del número a eliminar: "))
    try:
        numeros.pop(posicion)
        print("Número eliminado correctamente.")
        print(numeros)
    except IndexError:
        print("La posición indicada no existe.")

# 3. Función para visualizar un número según su posición
def visualizar_numero():
    posicion = int(input("Introduce la posición del número a visualizar: "))
    try:
        numero = numeros[posicion]
        print(f"El número en la posición {posicion} es: {numero}")
    except IndexError:
        print("La posición indicada no existe.")

# 4. Función para ordenar los números
def ordenar_numeros():
    numeros.sort()
    print("Lista de números ordenada correctamente.")
    print(numeros)

# 5. Función para visualizar la lista de números
def visualizar_lista():
    print("Lista de números:")
    for i, numero in enumerate(numeros):
        print(f"{i}: {numero}")

# 6. Función para añadir la lista al final del fichero c:\Ejercicios
def añadir_fichero():
    with open("C:\\Users\\alber\\OneDrive\\Escritorio\\Ejercicios", "a") as f:
        for numero in numeros:
            f.write(str(numero) + "\n")
    print("La lista se ha añadido correctamente al fichero:")
    print(f)
    
# Menú principal
while True:
    print("\nOpciones:")
    print("1. Añadir número a la lista")
    print("2. Eliminar número según su posición")
    print("3. Visualizar número según su posición")
    print("4. Ordernar los números")
    print("5. Visualizar lista")
    print("6. Añadir la lista al final del fichero c:/Ejercicios")
    print("7. Salir")
    opcion = int(input("Selecciona una opción: "))
    if opcion == 1:
        añadir_numero()
    elif opcion == 2:
        eliminar_numero()
    elif opcion == 3:
        visualizar_numero()
    elif opcion == 4:
        ordenar_numeros()
    elif opcion == 5:
        visualizar_lista()
    elif opcion == 6:
        añadir_fichero()
    elif opcion == 7:
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida.")
```

EJERCICIO 3

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n=1
estudiantes = {}
while(n<=10):
    nombre = input("Ingrese nombre: ")
    apellido = input("Ingrese apellido: ")
    email = input("Ingrese email: ")
    telefono = input("Ingrese telefono: ")
    nota = int(input("Ingrese nota: "))
    print("\n")

    estudiantes[n] = {'nombre':nombre, 'apellido':nota,'email':email,'telefono':telefono,'nota':nota}
    n+=1
    print(estudiantes)
##Estudiantes aprobados    
print("\nEstudiantes Aprobados")
aprobados=0
for n,d in estudiantes.items():
    if(d['nota']>=5):
        print(f"Nombre: {d['nombre']} Nota = {d['nota']}")
        aprobados=aprobados + 1
print("El número total de aprobados es:" + str(aprobados))
##Estudiantes suspendidos    
print("\nEstudiantes Suspendidos")
suspensos=0
for n,d in estudiantes.items():
    if(d['nota'] < 5):
        print(f"Nombre: {d['nombre']} Nota = {d['nota']}")
        suspensos=suspensos + 1
print("El número total de suspensos es:" + str(suspensos))

#Definimos los datos
x1=["aprobados"]
x2=["suspensos"]
y1 = [aprobados]
y2 = [suspensos]
#Configurar las características del gráfico
plt.bar(x1, y1, label = 'aprobados', width = 0.5, color = 'lightblue')
plt.bar(x2, y2, label = 'suspensos', width = 0.5, color = 'orange')
#Definir título y nombres de ejes
plt.title('Número de aprobados y suspensos')
plt.ylabel('Total de alumnos')
#Mostrar leyenda y figura
plt.legend()
plt.show()
```

```python
def main():
    numero = int(input("Dígame cuántas palabras tiene la lista: "))
    if numero < 1:
        print("¡Imposible!")
    else:
        lista = []
        for i in range(numero):
            palabra = input(f"Dígame la palabra {i + 1}: ")
            lista += [palabra]
            print(f"La lista creada es: {lista}")
            
    palabras = lista
    palabras_invertidas = []

    for palabra in palabras:
        palabra_invertida = palabra[::-1]
        palabras_invertidas.append(palabra_invertida)
        print(palabras_invertidas) 
             
if __name__ == "__main__":
  main()
```
